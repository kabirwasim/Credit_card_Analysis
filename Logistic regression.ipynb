```python
## Finding the right cut off 
```

```python
roc_df=pd.DataFrame()
for cutoff in np.linspace(0,1,50):
    train_data['pred_class']=np.where(train_data['pred']>cutoff,1,0)
    TP=np.where((train_data.default ==1) & (train_data.pred_class ==1),1,0)
    TN=np.where((train_data.default ==0) & (train_data.pred_class ==0),1,0)
    FP=np.where((train_data.default ==0) & (train_data.pred_class ==1),1,0)
    FN=np.where((train_data.default ==1) & (train_data.pred_class ==0),1,0)
    actual_1=(TP.sum()+FN.sum())
    actual_0=(TN.sum()+FP.sum())
    #sensitivity is recall of 1s
    recall=TP.sum()/actual_1 # recall is also known as sensitivity
    #specificity - is recall of 0s
    specificity=TN.sum()/actual_0
    total=(TP.sum()/actual_1)+(TN.sum()/actual_0)
    temp_df=pd.DataFrame([cutoff,recall,specificity,total]).T
    roc_df=pd.concat([roc_df,temp_df],axis=0)
```

```python
roc_df.columns=['cutoff','recall','specificity','total']
roc_df.sort_values(by='total',ascending=False)
```

```python
## After running this code, I find .2 is right cut off
```
